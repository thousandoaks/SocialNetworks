# 1. Measures of Centrality

## Degree Centrality


```python
G = nx.krackhardt_kite_graph()
```

```python
nx.draw(G,with_labels=True)
```

```python
print("Degree centrality")
degreeCentrality = nx.degree_centrality(G)
for v in G.nodes():
    print("%0.2d %5.3f" % (v, degreeCentrality[v]))
```

## EigenVector Centrality

```python
print("EigenVector centrality")
eigenvectorCentrality = nx.eigenvector_centrality(G)
for v in G.nodes():
    print("%0.2d %5.3f" % (v, eigenvectorCentrality[v]))
```

## PageRank Centrality

```python
print("PageRank centrality")
pageRankCentrality = nx.pagerank(G, alpha=0.9)
for v in G.nodes():
    print("%0.2d %5.3f" % (v, pageRankCentrality[v]))

```

```python
I = nx.path_graph(4)
nx.draw(I,with_labels=True)
```

## Katz Centrality

```python
print("Katz centrality")
phi = (1+math.sqrt(5))/2.0 # largest eigenvalue of adj matrix

katzCentrality = nx.katz_centrality(I,1/phi-0.01)
for v in I.nodes():
    print("%0.2d %5.3f" % (v, katzCentrality[v]))
```

# 5. Homophily and assortative mixing

## Assortative mixing by enumerative characteristics

```python
G2=nx.Graph()
G2.add_nodes_from([0,1],color='Team1')
G2.add_nodes_from([0,4],color='Team1')
G2.add_nodes_from([0,5],color='Team1')

G2.add_nodes_from([2,3],color='Team2')
G2.add_nodes_from([2,6],color='Team2')

G2.add_edges_from([(0,1),(0,4),(0,5),(2,3),(2,6),(0,2)])
```

```python
nx.draw(G2,with_labels=True)
```

```python
print(nx.attribute_assortativity_coefficient(G2,'color'))
```

## Assortative mixing by scalar characteristics


```python
G3=nx.Graph()
G3.add_nodes_from([0,1],size=2)
G3.add_nodes_from([0,7],size=3)

G3.add_nodes_from([2,3],size=30)
G3.add_nodes_from([2,5],size=50)
G3.add_nodes_from([2,6],size=30)



G3.add_edges_from([(0,1),(2,3),(2,5),(2,6),(0,7)])
```

```python
nx.draw(G3,with_labels=True)
```

```python
print(nx.numeric_assortativity_coefficient(G3,'size'))
```
